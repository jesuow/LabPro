### 3.1 Transformación de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
# Escriba su código aquí
def nota_chilena(dataframe_in, columna):
      # Aplicar la fórmula para convertir las notas a la escala chilena
  dataframe_in[columna] = (dataframe_in[columna] * 0.06) + 1
    
    # Redondear los valores al segundo decimal
  dataframe_in[columna] = dataframe_in[columna].round(2)
    
  return dataframe_in
  
  pass
````

### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempeño, la Universidad les solicita que obtengan los alumnos quienes han reprobado el año y los que han tenido desempeño de destacado. Para esto se considerará que un alumno ha reprobado el año cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerará destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una función que permita obtener tres Dataframes:
- El primero deberá poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero será un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el año.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
#Escriba su código aquí
def reprobados_destacados(dataframe_in):
    #primer data frame s010 a1umn0s cuy0 pr0medi0 es men0r a 4,0
    boolean_mask= dataframe_in.loc[:,"promedio"]< 4.0
    reprobados= dataframe_in.loc[boolean_mask]

     #Crear una copia del DataFrame
    df = dataframe_in.copy()
    #Convertir la columna pr0medi0 a tipo float para asegurarse de que las comparaciones funcionen
    df["promedio"] = pd.to_numeric(df["promedio"], errors='coerce')
    df.loc[df["promedio"] < 4.0, "promedio"] = 'R'

    #Data Frame con los alumnos destacados ordenados de mayor a menor.
    destacados = dataframe_in[
        (dataframe_in['math score'] > 5.8) &
        (dataframe_in['reading score'] > 5.8) &
        (dataframe_in['writing score'] > 5.8)
    ]
    destacados = destacados.sort_values(by='promedio', ascending=False)

    return reprobados, df, destacados
````

````python
# Inserte su respuesta aquí
def resumen(dataframe_in):
    #calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
    dataframe_in["GPA"] = dataframe_in[["math score", "reading score", "writing score"]].mean(axis=1).round(2)

    #hacer una simplificación a través de un mapeo
    mapeo = {
        "some high school": "school",
        "some college": "school",
        "high school": "school",
        "bachelor's degree": "college",
        "associate's degree": "college",
        "master's degree": "postgraduate"
    }
    #los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.
    dataframe_in["simple parental level of education"] = dataframe_in["parental level of education"].map(mapeo)

    #promedio de las notas por 'race/ethnicity' y 'simple parental level of education'
    resumen = dataframe_in.groupby(["race/ethnicity", "simple parental level of education"])[["GPA", "math score", "reading score", "writing score"]].mean().reset_index()

    #contar la cantidad de alumnos por grupo
    conteo = dataframe_in.groupby(["race/ethnicity", "simple parental level of education"]).size().reset_index(name="count")

    #unir el promedio y el conteo
    resumen = pd.merge(resumen, conteo, on=["race/ethnicity", "simple parental level of education"])

    #porcentaje de alumnos
    total = resumen["count"].sum()
    resumen["percentage"] = (resumen["count"] / total * 100).round(2).astype(str) + "%"

    #redondear los valores de las notas a dos decimales
    resumen[["GPA", "math score", "reading score", "writing score"]] = resumen[["GPA", "math score", "reading score", "writing score"]].round(2)

    return resumen
````

El rector (evidentemente molesto por la situación) les ruega incluir estos datos (vaciados en el archivo other_grades.csv) al estudio original(`students_grades.csv`).


Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situación (y así evitar buscar duplicados o **alumnos que no pertenecen a la universidad**).

- [ ] Unir `df_grades` con `other_grades.csv` usando right join y explique el resultado.

> **Hint**: Puede explicar los resultados del merge a través de la cantidad de filas resultantes y los valores que estas contienen.

**Respuesta:**

````python
# Inserte su respuesta aquí
other_grades = pd.read_csv('other_grades.csv')

df_grades_outer = pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    sort=True,
    how="outer",
)
print("# filas con outer join:", df_grades_outer.shape[0])
print(df_grades_outer.head())


df_grades_left = pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    how="left",
    sort=True,)
print("# con left join:", df_grades_left.shape[0])
print(df_grades_left.head())

df_grades_right = pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    how="right",
    sort=True,)
print("# con right join:", df_grades_right.shape[0])
print(df_grades_right.head())

df_grades_inner = pd.merge(
    left=df_grades,
    right=other_grades,
    left_on="names",
    right_on="names",
    how="inner",
    sort=True,)
print("# con inner join:", df_grades_inner.shape[0])
print(df_grades_inner.head())
````


Outer Join: Combina todas las filas de ambos DataFrames. Si un alumno solo está en uno de los DataFrames, aparecerá la fila, pero las columnas del otro DataFrame estarán vacías para ese alumno. Resultado: 1150 filas, con valores nulos para alumnos que no coinciden en ambos datasets.

Left Join: Muestra todas las filas del DataFrame original (df_grades), añadiendo las columnas del DataFrame de las nuevas notas (other_grades). Si un alumno no está en other_grades, sus columnas de historia y ciencias quedarán vacías. Resultado: 1000 filas.

Right Join: Muestra todas las filas del DataFrame other_grades, agregando datos de df_grades solo para los alumnos que coinciden. Alumnos en other_grades que no están en df_grades aparecerán con valores nulos en las demás columnas. Resultado: 1150 filas.

Inner Join: Solo incluye los alumnos que están presentes en ambos DataFrames, asegurando que todos los valores están completos. Resultado: 1000 filas, sin valores nulos.

Si quiero incluir únicamente a los alumnos que están en ambos DataFrames, el inner join es la mejor opción. Esto garantiza que las notas de ambas asignaturas (historia y ciencias) estén disponibles, eliminando cualquier registro que no las tenga y evitando inconsistencias en el análisis.

Sin embargo, si prefiero conservar a todos los alumnos del DataFrame original y solo agregar las notas de las nuevas asignaturas donde estén disponibles, el left join sería la opción más adecuada. Dado que ambos métodos me devuelven 1000 filas, en este caso no hay diferencia significativa entre usar uno u otro, por lo que cualquiera de los dos sería válido.
