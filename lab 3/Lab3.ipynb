### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempe√±o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a√±o y los que han tenido desempe√±o de destacado. Para esto se considerar√° que un alumno ha reprobado el a√±o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar√° destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci√≥n que permita obtener tres Dataframes:
- El primero deber√° poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser√° un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a√±o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
#Escriba su c√≥digo aqu√≠
def reprobados_destacados(dataframe_in):
    #primer data frame s010 a1umn0s cuy0 pr0medi0 es men0r a 4,0
    boolean_mask= dataframe_in.loc[:,"promedio"]< 4.0
    reprobados= dataframe_in.loc[boolean_mask]

     #Crear una copia del DataFrame
    df = dataframe_in.copy()
    #Convertir la columna pr0medi0 a tipo float para asegurarse de que las comparaciones funcionen
    df["promedio"] = pd.to_numeric(df["promedio"], errors='coerce')
    df.loc[df["promedio"] < 4.0, "promedio"] = 'R'

    #Data Frame con los alumnos destacados ordenados de mayor a menor.
    destacados = dataframe_in[
        (dataframe_in['math score'] > 5.8) &
        (dataframe_in['reading score'] > 5.8) &
        (dataframe_in['writing score'] > 5.8)
    ]
    destacados = destacados.sort_values(by='promedio', ascending=False)

    return reprobados, df, destacados
````

        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar seg√∫n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [ ] Agregar seg√∫n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.

Finalmente, responda las siguientes preguntas sobre el gr√°fico de caja que viene implementado a continuaci√≥n. En este se muestra el GPA con respecto al nivel educacional, donde la variable de color es la etnicidad.

- [ ]  ¬øHay alguna diferencia entre los grupos graficados tanto para el nivel educacional de los padres como tambi√©n para la etnicidad?


**Respuesta:**

````python
# Inserte su respuesta aqu√≠
def resumen(dataframe_in):
    #calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
    dataframe_in["GPA"] = dataframe_in[["math score", "reading score", "writing score"]].mean(axis=1).round(2)

    #hacer una simplificaci√≥n a trav√©s de un mapeo
    mapeo = {
        "some high school": "school",
        "some college": "school",
        "high school": "school",
        "bachelor's degree": "college",
        "associate's degree": "college",
        "master's degree": "postgraduate"
    }
    #los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.
    dataframe_in["simple parental level of education"] = dataframe_in["parental level of education"].map(mapeo)

    #promedio de las notas por 'race/ethnicity' y 'simple parental level of education'
    resumen = dataframe_in.groupby(["race/ethnicity", "simple parental level of education"])[["GPA", "math score", "reading score", "writing score"]].mean().reset_index()

    #contar la cantidad de alumnos por grupo
    conteo = dataframe_in.groupby(["race/ethnicity", "simple parental level of education"]).size().reset_index(name="count")

    #unir el promedio y el conteo
    resumen = pd.merge(resumen, conteo, on=["race/ethnicity", "simple parental level of education"])

    #porcentaje de alumnos
    total = resumen["count"].sum()
    resumen["percentage"] = (resumen["count"] / total * 100).round(2).astype(str) + "%"

    #redondear los valores de las notas a dos decimales
    resumen[["GPA", "math score", "reading score", "writing score"]] = resumen[["GPA", "math score", "reading score", "writing score"]].round(2)

    return resumen
````

El rector (evidentemente molesto por la situaci√≥n) les ruega incluir estos datos (vaciados en el archivo other_grades.csv) al estudio original(`students_grades.csv`).


Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situaci√≥n (y as√≠ evitar buscar duplicados o **alumnos que no pertenecen a la universidad**).

> **Hint**: Puede explicar los resultados del merge a trav√©s de la cantidad de filas resultantes y los valores que estas contienen.

**Respuesta:**

````python
# Inserte su respuesta aqu√≠
other_grades = pd.read_csv('other_grades.csv')

df_grades_outer = pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    sort=True,
    how="outer",
)
print("# filas con outer join:", df_grades_outer.shape[0])
print(df_grades_outer.head())


df_grades_left = pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    how="left",
    sort=True,)
print("# con left join:", df_grades_left.shape[0])
print(df_grades_left.head())

df_grades_right = pd.merge(
    left=df_grades,
    right=other_grades,
    on="names",
    how="right",
    sort=True,)
print("# con right join:", df_grades_right.shape[0])
print(df_grades_right.head())

df_grades_inner = pd.merge(
    left=df_grades,
    right=other_grades,
    left_on="names",
    right_on="names",
    how="inner",
    sort=True,)
print("# con inner join:", df_grades_inner.shape[0])
print(df_grades_inner.head())
````


Outer Join: Combina todas las filas de ambos DataFrames. Si un alumno solo est√° en uno de los DataFrames, aparecer√° la fila, pero las columnas del otro DataFrame estar√°n vac√≠as para ese alumno. Resultado: 1150 filas, con valores nulos para alumnos que no coinciden en ambos datasets.

Left Join: Muestra todas las filas del DataFrame original (df_grades), a√±adiendo las columnas del DataFrame de las nuevas notas (other_grades). Si un alumno no est√° en other_grades, sus columnas de historia y ciencias quedar√°n vac√≠as. Resultado: 1000 filas.

Right Join: Muestra todas las filas del DataFrame other_grades, agregando datos de df_grades solo para los alumnos que coinciden. Alumnos en other_grades que no est√°n en df_grades aparecer√°n con valores nulos en las dem√°s columnas. Resultado: 1150 filas.

Inner Join: Solo incluye los alumnos que est√°n presentes en ambos DataFrames, asegurando que todos los valores est√°n completos. Resultado: 1000 filas, sin valores nulos.

Si quiero incluir √∫nicamente a los alumnos que est√°n en ambos DataFrames, el inner join es la mejor opci√≥n. Esto garantiza que las notas de ambas asignaturas (historia y ciencias) est√©n disponibles, eliminando cualquier registro que no las tenga y evitando inconsistencias en el an√°lisis.

Sin embargo, si prefiero conservar a todos los alumnos del DataFrame original y solo agregar las notas de las nuevas asignaturas donde est√©n disponibles, el left join ser√≠a la opci√≥n m√°s adecuada. Dado que ambos m√©todos me devuelven 1000 filas, en este caso no hay diferencia significativa entre usar uno u otro, por lo que cualquiera de los dos ser√≠a v√°lido.
